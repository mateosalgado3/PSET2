```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

DATA_PATH = "/mnt/c/Users/pmate/OneDrive - Universidad San Francisco de Quito/8 SEMESTRE/DataMining/Deberes/PSet2/data/processed/Airbnb_Featured.csv"

df = pd.read_csv(DATA_PATH)

display(df.head())

#  Verificar distribución de la nueva variable "total_amenities"
plt.figure(figsize=(8,5))
sns.histplot(df["total_amenities"], bins=30, kde=True)
plt.title("Distribución de Total de Amenidades")
plt.xlabel("Número de Amenidades")
plt.ylabel("Frecuencia")
plt.show()

# Comparar antes y después del escalado
num_features = ["accommodates", "bathrooms", "bedrooms", "beds", "number_of_reviews", "review_scores_rating", "total_amenities"]

fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten()

for i, col in enumerate(num_features):
    sns.histplot(df[col], bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f"Distribución de {col}")

plt.tight_layout()
plt.show()

# Filtrar solo las columnas numéricas para evitar errores con strings
numeric_df = df.select_dtypes(include=[np.number])

# Calcular la matriz de correlación
plt.figure(figsize=(12,6))
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Matriz de Correlación (Solo Variables Numéricas)")
plt.show()
```
